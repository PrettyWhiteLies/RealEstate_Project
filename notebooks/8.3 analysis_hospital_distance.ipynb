# Analysis Hospital Distance

```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
folder_path = '../data/raw/hospital_distance/'

combined_df = pd.DataFrame()

file_list = os.listdir(folder_path)

for file_name in file_list:
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
combined_df.to_csv('../data/raw/merge_requirement/Driving_hospital.csv', index=False)
```

## Scatter Plot of Nearest Hospital Station Distance vs Price

```python
plt.figure(figsize=(8,6))
plt.scatter(combined_df['price'], combined_df['NearestHospitalDistance'], alpha=0.5) 
plt.xlabel('price')
plt.ylabel('Nearest Hospital Distance')
plt.title('Scatter Plot of Nearest Hospital Distance vs Price')
plt.savefig('../plots/Scatter Plot of Nearest Hospital Distance vs Price.png', dpi=300)
plt.show()
```

## Box plot whether nearest Hospital station is within 1 km

```python
combined_df['NearestHospital'] = combined_df['NearestHospitalDistance'] > 1
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x='NearestHospital', y='price', data=combined_df)
plt.xlabel('NearestHospital (NearestHospitalDistance > 1)')
plt.ylabel('Price')
plt.title('Box Plot of Price vs. NearestHospital')
plt.savefig('../plots/boxplot_hospital.png', dpi=300)
plt.show()
```
