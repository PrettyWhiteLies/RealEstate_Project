# Analysis Train Distance

```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/merge_requirement/Direct_train.csv')
```

## Log Scatter Plot of Nearest Train Station Distance vs. Rental Price

```python
plt.figure(figsize=(8,6))
plt.scatter(np.log(df['price']), np.log(df['NearestTrainDistance']), alpha=0.5) 

plt.xlabel('Log Price')
plt.ylabel('Log Nearest Train Distance')
plt.title('Scatter Plot of Log-Transformed Nearest Train Distance vs. Log-Transformed Price')

plt.savefig('../plots/Scatter Plot of Log-Transformed Nearest Train Distance vs. Log-Transformed Price.png')

plt.show()
```

## Box Plot whether nearest train station is within 1 km

```python
df['NearestTrain'] = df['NearestTrainDistance'] > 1


sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='NearestTrain', y='price', data=df)


plt.xlabel('NearestTrain (NearestTrainDistance > 1)')
plt.ylabel('Price')
plt.title('Box Plot of Price vs. NearestTrain')

plt.savefig('../plots/boxplot_train.png', dpi=300)

plt.show()
```
