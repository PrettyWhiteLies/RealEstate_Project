# Analysis School Distance

```python
import pandas as pd
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/merge_requirement/Direct_school.csv')
```

## Log Scatter Plot of Nearest School Distance vs. Rental Price

To make the graph more clear, we use log transformation to reduce the influence of extreme values

```python
plt.figure(figsize=(8, 6))
plt.scatter(np.log(df['price']), np.log(df['NearestSchoolDistance']), alpha=0.5)

plt.xlabel('Log Price')
plt.ylabel('Log Nearest School Distance')
plt.title('Scatter Plot of Log-Transformed Nearest School Distance vs. Log-Transformed Price')

plt.savefig('../plots/Scatter Plot of Log-Transformed Nearest School Distance vs Log-Transformed Price.png')

plt.show()
```

## Box Plot whether nearest school is within 4 km

```python
df['NearestSchool'] = df['NearestSchoolDistance'] > 4

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='NearestSchool', y='price', data=df)

plt.xlabel('NearestSchool (NearestSchoolDistance > 1)')
plt.ylabel('Price')
plt.title('Box Plot of Price vs. NearestSchool')

plt.savefig('../plots/boxplot_school.png', dpi=300)

plt.show()
```
