# Analysis Park Distance

```python
import pandas as pd
import os
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/merge_requirement/Direct_park.csv')
```

## Log Scatter Plot of Nearest Park Distance vs. Rental Price

```python
plt.figure(figsize=(8, 6))

# Apply a logarithmic transformation to the data
log_price = np.log(df['price'])
log_park_distance = np.log(df['NearestParkDistance'])

plt.scatter(log_price, log_park_distance, alpha=0.5)

plt.xlabel('Log Price')
plt.ylabel('Log Nearest Park Distance')
plt.title('Scatter Plot of Log-Transformed Nearest Park Distance vs. Log-Transformed Price')

plt.savefig('../plots/Scatter Plot of Log-Transformed Nearest Park Distance vs. Log-Transformed Price.png')

plt.show()
```

## Box Plot whether nearest Park station is within 2 km

```python
df['NearestPark'] = df['NearestParkDistance'] > 2


sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.boxplot(x='NearestPark', y='price', data=df)


plt.xlabel('NearestPark (NearestParkDistance > 2)')
plt.ylabel('Price')
plt.title('Box Plot of Price vs. NearestPark')

plt.savefig('../plots/boxplot_park.png', dpi=300)

plt.show()
```
